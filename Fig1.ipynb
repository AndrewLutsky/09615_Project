```python
import pandas as pd

import seaborn as sns

import matplotlib.pyplot as plt



# Results dictionary

results = {

    "Model": ["Logistic Regression", "Logistic Regression", "Decision Tree", "Decision Tree",

              "Random Forest", "Random Forest", "XGBoost", "XGBoost"],

    "Metric": ["F1 Score", "Accuracy", "F1 Score", "Accuracy", 

               "F1 Score", "Accuracy", "F1 Score", "Accuracy"],

    "Value": [

        (0.7480 + 0.7531 + 0.2538 + 0.2390) / 4,  # Logistic Regression F1

        (0.7437 + 0.7484 + 0.8659 + 0.8650) / 4,  # Logistic Regression Accuracy

        (0.6544 + 0.6487 + 0.3150 + 0.2842) / 4,  # Decision Tree F1

        (0.6681 + 0.6493 + 0.7984 + 0.8323) / 4,  # Decision Tree Accuracy

        (0.7126 + 0.7474 + 0.2561 + 0.2939) / 4,  # Random Forest F1

        (0.7062 + 0.7375 + 0.8607 + 0.8446) / 4,  # Random Forest Accuracy

        (0.7608 + 0.7628 + 0.2662 + 0.2570) / 4,  # XGBoost F1

        (0.7505 + 0.7536 + 0.8675 + 0.8673) / 4   # XGBoost Accuracy

    ]

}



# Convert to DataFrame

df = pd.DataFrame(results)

# Create a seaborn bar plot

plt.figure(figsize=(8, 8))

sns.barplot(data=df, x="Model", y="Value", hue="Metric")

plt.title("Average Model Performance Across All Datasets")

plt.ylabel("Average Metric Value")

plt.xlabel("Model")

plt.legend(title="Metric")

plt.tight_layout()



# Display the plot

plt.show()


```
