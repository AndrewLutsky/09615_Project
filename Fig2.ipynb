```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Data Preparation
data = {
    "Dataset": [
        "Balanced Filtered", "Balanced Filtered", "Balanced Filtered", "Balanced Filtered",
        "Balanced Full", "Balanced Full", "Balanced Full", "Balanced Full",
        "Unbalanced Full", "Unbalanced Full", "Unbalanced Full", "Unbalanced Full",
        "Unbalanced Filtered", "Unbalanced Filtered", "Unbalanced Filtered", "Unbalanced Filtered"
    ],
    "F1 Score": [
        0.7460, 0.7495, 0.7578, 0.7607,
        0.7531, 0.6487, 0.7474, 0.7628,
        0.2538, 0.3150, 0.2561, 0.2662,
        0.2390, 0.2842, 0.2939, None
    ],
    "Accuracy": [
        0.7460, 0.7413, 0.7492, 0.7512,
        0.7484, 0.6493, 0.7375, 0.7536,
        0.8659, 0.7984, 0.8607, 0.8675,
        0.8650, 0.8323, None, None
    ]
}

# Convert to DataFrame
df = pd.DataFrame(data)

# Calculate averages grouped by dataset type
average_metrics = df.groupby("Dataset")[["F1 Score", "Accuracy"]].mean().reset_index()

# Plot using Seaborn
plt.figure(figsize=(10, 6))
average_metrics = pd.melt(
    average_metrics, id_vars=["Dataset"], value_vars=["F1 Score", "Accuracy"], var_name="Metric", value_name="Value"
)

sns.barplot(
    data=average_metrics,
    x="Dataset",
    y="Value",
    hue="Metric",
    edgecolor="black", 
    linewidth=1.5
)

# Add labels and title
plt.title("Average F1 Score and Accuracy for Balanced vs Unbalanced Datasets", fontsize=16)
plt.xlabel("Dataset", fontsize=14)
plt.ylabel("Average Value", fontsize=14)
plt.legend(title="Metric", fontsize=12, title_fontsize=14)
plt.ylim(0, 1)  # Metrics are between 0 and 1 for better visualization
plt.figure(figsize=(8, 8))

# Show the plot
plt.tight_layout()
plt.show()
```
